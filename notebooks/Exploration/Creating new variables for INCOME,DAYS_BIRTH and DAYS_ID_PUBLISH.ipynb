```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('application_data.csv')
df
```

```python
df.describe()
```

```python
object_cols = df.select_dtypes(include=['object'])
numeric_cols = df.select_dtypes(include=['int64', 'float64']).iloc[:, :13]

# Concatenate both selections
selected_columns = pd.concat([object_cols, numeric_cols], axis=1)

# Slice the DataFrame using the selected columns
data = df[selected_columns.columns]

data
```

```python
description = data.describe(include='all')
description
```

```python
data.dtypes
```

```python
corr = data.corr(numeric_only=True)

mask = np.triu(np.ones_like(corr, dtype=bool))

plt.figure(figsize=(10, 8),dpi=100)

cmap = sns.diverging_palette(230, 20, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True, fmt=".2f")

plt.title('Correlation Heatmap', size=12)

plt.show()
```

```python
sns.boxplot(data=data['AMT_INCOME_TOTAL'])

# Add title and labels
plt.title('Boxplot of Column')
plt.xlabel('Column Name')
plt.ylabel('Values')

# Show the plot
plt.show()
print("The figure shows AMT_INCOME_TOTAL column has so many outliers. " )
```

```python
corr = data.corr(numeric_only=True)

mask = np.triu(np.ones_like(corr, dtype=bool))
```

```python
#Since AMT_INCOME_TOTAL value have more outlier we can use logarithmic function to reduce the effect of outliers 
data.loc[:, 'Income_squared'] = data['AMT_INCOME_TOTAL'].apply(lambda x: np.log(x + 1))

#To get the interaction effect DAYS_BIRTH and DAYS_ID_PUBLISH are multiplied.
def create_new_feature(row):
    # Example: Concatenate values from two existing columns to create a new feature
    new_feature_value = row['DAYS_BIRTH'] * row['DAYS_ID_PUBLISH']
    return new_feature_value

# Apply the function to create the new feature for each row in the DataFrame
data.loc[:, 'Birthdays_Publishdays'] = data.apply(create_new_feature, axis=1)
```

```python
corr = data.corr(numeric_only=True)

mask = np.triu(np.ones_like(corr, dtype=bool))

plt.figure(figsize=(10, 8),dpi=100)

cmap = sns.diverging_palette(230, 20, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True, fmt=".2f")

plt.title('Correlation Heatmap', size=12)

plt.show()

print("By creating new variables the correlation with target became -0.07 and -0.02 for product of DAYS_BIRTH and DAYS_ID_PUBLISH and squared income respectively " )
```
