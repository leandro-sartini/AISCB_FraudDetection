```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#corr=df.corr(numeric_only=True)
corr = df.corr(numeric_only=True)

mask = np.triu(np.ones_like(corr, dtype=bool))
```

```python
plt.figure(figsize=(20,10),dpi=100)

cmap = sns.diverging_palette(230, 20, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True, fmt=".2f")

plt.title('Correlation Chart', size=12)

plt.show()
```

```python
plt.figure(figsize=[10,8],dpi= 100)

sns.scatterplot(df,x='CNT_CHILDREN',y='CNT_FAM_MEMBERS',color='#FFB5E8')

plt.title('CNT_CHILDREN VS CNT_FAM_MEMBERS', size=12)

plt.show()
print("From the correlation chart we can understand that 'CNT_CHILDREN' and 'CNT_FAM_MEMBERS' are highly correlated")

```
