```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('application_data.csv')
df
```

```python
df.describe()
```

```python
object_cols = df.select_dtypes(include=['object'])
numeric_cols = df.select_dtypes(include=['int64', 'float64']).iloc[:, :13]

# Concatenate both selections
selected_columns = pd.concat([object_cols, numeric_cols], axis=1)

# Slice the DataFrame using the selected columns
data = df[selected_columns.columns]

data
```

```python
description = data.describe(include='all')
description
```

```python
data.dtypes
```

```python
corr = data.corr(numeric_only=True)

mask = np.triu(np.ones_like(corr, dtype=bool))

plt.figure(figsize=(10, 8),dpi=100)

cmap = sns.diverging_palette(230, 20, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True, fmt=".2f")

plt.title('Correlation Heatmap', size=12)

plt.show()
```

```python
column_name = 'CNT_CHILDREN'

# Calculate skewness of the column count of children
skewness = df[column_name].skew()

# Print the skewness value
print("Skewness of", column_name, ":", skewness)

# Determine if the data is skewed
if skewness > 1 or skewness < -1:
    print(column_name, "data is highly skewed.")
elif 0.5 < skewness < 1 or -1 < skewness < -0.5:
    print(column_name, "data is moderately skewed.")
else:
    print(column_name, "data is approximately symmetrically distributed.")
```

```python
# Compute square root of 'REGION_POPULATION_RELATIVE' and create a new column
data.loc[:, 'SquareRoot_cnt_children'] = np.sqrt(data['REGION_POPULATION_RELATIVE'])

# Compute product of 'REGION_POPULATION_RELATIVE' and 'AMT_GOODS_PRICE' and create a new column
data.loc[:, 'Region_Goods_price'] = data['REGION_POPULATION_RELATIVE'] * data['AMT_GOODS_PRICE']
```

```python
column_name = 'SquareRoot_cnt_children'

# Calculate skewness of the column count of children
skewness = data[column_name].skew()

# Print the skewness value
print("Skewness of", column_name, ":", skewness)

# Determine if the data is skewed
if skewness > 1 or skewness < -1:
    print(column_name, "data is highly skewed.")
elif 0.5 < skewness < 1 or -1 < skewness < -0.5:
    print(column_name, "data is moderately skewed.")
else:
    print(column_name, "data is approximately symmetrically distributed.")
```

```python
corr = data.corr(numeric_only=True)

mask = np.triu(np.ones_like(corr, dtype=bool))

plt.figure(figsize=(10, 8),dpi=100)

cmap = sns.diverging_palette(230, 20, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True, fmt=".2f")

plt.title('Correlation Heatmap', size=12)

plt.show()

print("By creating new variable SquareRoot_cnt_children the correlation with target became -0.03 and for product of REGION_POPULATION_RELATIVE and AMT_GOODS_PRICE gives -0.05 correlation" )
```
