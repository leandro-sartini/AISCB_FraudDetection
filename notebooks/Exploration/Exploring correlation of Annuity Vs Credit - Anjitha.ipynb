```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('application_data.csv')
df
```

```python
df.describe()
```

# Select only 30 columns from dataset which are comparitively affecting the target value  

```python
object_cols = df.select_dtypes(include=['object'])
numeric_cols = df.select_dtypes(include=['int64', 'float64']).iloc[:, :13]

# Concatenate both selections
selected_columns = pd.concat([object_cols, numeric_cols], axis=1)

# Slice the DataFrame using the selected columns
data = df[selected_columns.columns]
data
```

```python
data.describe()
```

```python
data.dtypes
```

# Correlation chart

```python
correlation = data.corr(numeric_only=True)
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(correlation, annot=True, cmap='hot', fmt=".2f")
plt.title('Correlation Chart')
plt.show()
```

# Plotting Multicolinearity

```python
plt.scatter(df['AMT_CREDIT'], df['AMT_GOODS_PRICE'])
plt.xlabel('AMT_CREDIT')
plt.ylabel('AMT_GOODS_PRICE')
plt.show()
print("From the correlation chart we can understand AMT_CREDIT and AMT_GOODS_PRICE are the highly correlated features \nWhen AMT_GOODS_PRICE increases AMT_CREDIT is also going high " )
```

```python
plt.scatter(df['AMT_GOODS_PRICE'], df['AMT_ANNUITY'])
plt.xlabel('AMT_GOODS_PRICE')
plt.ylabel('AMT_ANNUITY')
plt.show()
print("From the correlation chart we can understand AMT_GOODS_PRICE and AMT_ANNUITY are correlated.\nAMT_ANNUITY increasing with increase in AMT_GOODS_PRICE" )
```

```python
plt.scatter(df['AMT_CREDIT'], df['AMT_ANNUITY'])
plt.xlabel('AMT_CREDIT')
plt.ylabel('AMT_ANNUITY')
plt.show()
print("From the correlation chart we can understand AMT_CREDIT and AMT_ANNUITY are correlated.\nAMT_ANNUITY increasing with increase in AMT_CREDIT" )
```

```python
plt.scatter(df['DAYS_EMPLOYED'], df['DAYS_BIRTH'])
plt.xlabel('DAYS_EMPLOYED')
plt.ylabel('DAYS_BIRTH')
plt.show()
print("From the correlation chart we can understand DAYS_EMPLOYED and DAYS_BIRTH are negatively correlated." )
```

```python
plt.scatter(df['DAYS_ID_PUBLISH'], df['DAYS_EMPLOYED'])
plt.xlabel('DAYS_ID_PUBLISH')
plt.ylabel('DAYS_EMPLOYED')
plt.show()
print("From the correlation chart we can understand DAYS_EMPLOYED and DAYS_ID_PUBLISH are negatively correlated." )
```

```python
plt.scatter(df['CNT_CHILDREN'], df['DAYS_EMPLOYED'])
plt.xlabel('CNT_CHILDREN')
plt.ylabel('DAYS_EMPLOYED')
plt.show()
print("From the correlation chart we can understand DAYS_EMPLOYED and CNT_CHILDREN are negatively correlated." )
```
