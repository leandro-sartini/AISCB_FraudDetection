```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('application_data.csv')
df
```

```python
df.describe()
```

# Select only 30 columns from dataset which are comparitively affecting the target value  

```python
object_cols = df.select_dtypes(include=['object'])
numeric_cols = df.select_dtypes(include=['int64', 'float64']).iloc[:, :13]

# Concatenate both selections
selected_columns = pd.concat([object_cols, numeric_cols], axis=1)

# Slice the DataFrame using the selected columns
data = df[selected_columns.columns]

data
```

```python
data.describe()
```

```python
data.dtypes
```

# Correlation chart

```python
corr = data.corr(numeric_only=True)

mask = np.triu(np.ones_like(corr, dtype=bool))
```

```python
plt.figure(figsize=(10, 8),dpi=100)

cmap = sns.diverging_palette(230, 20, as_cmap=True)

sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True, fmt=".2f")

plt.title('Correlation Heatmap', size=12)

plt.show()
```

# Plotting Multicolinearity

```python
plt.figure(figsize=[10,8],dpi= 100)

sns.scatterplot(data,x='AMT_CREDIT',y='AMT_GOODS_PRICE',color='#FFB5E8')

plt.title('AMT_CREDIT VS AMT_GOODS_PRICE', size=12)

plt.show()

print("From the correlation chart we can understand AMT_CREDIT and AMT_GOODS_PRICE are the highly correlated features \nWhen AMT_GOODS_PRICE increases AMT_CREDIT is also going high " )
```

```python
plt.figure(figsize=[10,8],dpi= 100)

sns.scatterplot(data,x='AMT_GOODS_PRICE',y='AMT_ANNUITY',color='#FFB5E8')

plt.title('AMT_GOODS_PRICE VS AMT_ANNUITY', size=12)

plt.show()

print("From the correlation chart we can understand AMT_GOODS_PRICE and AMT_ANNUITY are correlated.\nAMT_ANNUITY increasing with increase in AMT_GOODS_PRICE" )
```

```python
plt.figure(figsize=[10,8],dpi= 100)

sns.scatterplot(data,x='AMT_CREDIT',y='AMT_ANNUITY',color='#FFB5E8')

plt.title('AMT_CREDIT VS AMT_ANNUITY', size=12)


plt.show()

print("From the correlation chart we can understand AMT_CREDIT and AMT_ANNUITY are correlated.\nAMT_ANNUITY increasing with increase in AMT_CREDIT" )
```

```python
plt.figure(figsize=[10,8],dpi= 100)

sns.scatterplot(data,x='DAYS_EMPLOYED',y='DAYS_BIRTH', color='#FFB5E8')

plt.title('DAYS_EMPLOYED VS DAYS_BIRTH', size=12)

plt.show()

print("From the correlation chart we can understand DAYS_EMPLOYED and DAYS_BIRTH are negatively correlated." )
```

# Plotting higly correlated features with target

```python
fig, axs = plt.subplots(nrows=1,ncols=2,figsize=[10,4],dpi= 120)

sns.boxplot(data,x='TARGET', y='AMT_CREDIT',ax=axs[0],palette={0:"green",1:"red"})
axs[0].set_title('TARGET VS AMT_CREDIT')

sns.boxplot(data,x='TARGET', y='AMT_GOODS_PRICE',ax=axs[1], palette={0:"green",1:"red"})
axs[1].set_title('TARGET VS AMT_GOODS_PRICE')

plt.show()

print('Both AMT_CREDIT and AMT_GOODS_PRICE features are correlated to Target on almost same level\nSo we can drop AMT_GOODS_PRICE feature')
```
